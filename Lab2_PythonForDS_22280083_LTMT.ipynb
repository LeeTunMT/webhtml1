## Question 1: The probability of passing three candidates A, B, and C in one interview are 60%, 10%, and 20% respectively. Find the probability of success of at least two

### Probability of Passing Candidates

The probabilities of passing for candidates A, B, and C in one interview are 60%, 10%, and 20%, respectively. We need to find the probability that at least two candidates pass.

Let:  
- \( P(A) = 0.6 \)  
- \( P(B) = 0.1 \)  
- \( P(C) = 0.2 \)

1. **Probability that no candidates pass (\( P(X = 0) \))**:  
   $$ P(X = 0) = (1 - P(A)) \times (1 - P(B)) \times (1 - P(C)) $$  
   $$ = 0.4 \times 0.9 \times 0.8 = 0.288 $$

2. **Probability that only one candidate passes (\( P(X = 1) \))**:  
   $$ P(X = 1) = P(A) \times (1 - P(B)) \times (1 - P(C)) $$  
   $$ + P(B) \times (1 - P(A)) \times (1 - P(C)) $$  
   $$ + P(C) \times (1 - P(A)) \times (1 - P(B)) $$  
   $$ = 0.6 \times 0.9 \times 0.8 + 0.1 \times 0.4 \times 0.8 + 0.2 \times 0.4 \times 0.9 = 0.536 $$

3. **Probability that at least two candidates pass (\( P(X \geq 2) \))**:  
   $$ P(X \geq 2) = 1 - P(X = 0) - P(X = 1) $$  
   $$ = 1 - 0.288 - 0.536 = 0.176 $$

Therefore, the probability that at least two candidates pass is **0.176** or **17.6%**.

   $$ \text{Total Ways} = 1 x 12C4 x 12 x 45C2 = 5880600 $$

   - There are 12 ways to choose a heart from the remaining 12 hearts.
   - There are 2 * 44 - 1 = 87 ways to draw other 44 cards, ensuring at least one is a king from the spades or clubs.

   $$ \text{Total Ways} = 1 \times \binom{12}{5} \times 12 \times 87 $$

3. **Case 3: Diamond King Was Drawn, but Heart King Was Not**

4. **Case 4: Neither Heart King nor Diamond King Were Drawn**  
   - There are 12C5 ways to choose 5 diamonds from 12 diamonds (excluding the diamond king).
   - There are 12C5 ways to choose 2 hearts from 12 hearts (excluding the heart king).
   - There is 1 way to choose the other 2 kings.

   $$ \text{Total Ways} = \binom{12}{5} \times \binom{12}{2} $$

### Probability of Fraud Detection in Transactions from Banks A, B, and C

There are transactions from banks A, B, and C at your company. Approximately 30% of the transactions come from bank A, 20% from bank B, and 50% from bank C. It is known that there is a possibility of fraud in the transaction history for these banks. The fraud rates for banks A, B, and C are 3%, 5%, and 4%, respectively. We want to determine which bank is most likely involved when a fraud is detected.

Assuming there is a total of 1000 transactions at the company, we have:
- 300 transactions from bank A
- 200 transactions from bank B
- 500 transactions from bank C

Let \( P_A \), \( P_B \), and \( P_C \) be the probabilities of fraud for transactions from banks A, B, and C, respectively:
- \( P_A = 0.03 \)
- \( P_B = 0.05 \)
- \( P_C = 0.04 \)

The theoretical number of fraudulent transactions for each bank is calculated as follows:

1. **Bank A**:  
   $$ |A| = 300 \times P_A = 300 \times 0.03 = 9 \text{ frauds} $$

2. **Bank B**:  
   $$ |B| = 200 \times P_B = 200 \times 0.05 = 10 \text{ frauds} $$

3. **Bank C**:  
   $$ |C| = 500 \times P_C = 500 \times 0.04 = 20 \text{ frauds} $$

Comparing the theoretical number of frauds for each bank, we see that bank C has the highest number of fraudulent transactions (20), which is significantly larger than those of bank A (9) and bank B (10).

### Conclusion

Based on this analysis, **bank C** is the most likely to be involved when a fraud is detected.

### My ideas and ways how to find out Outliersand the ways to treat them

1. **Calculate Q1 and Q3**: Find the 25th percentile (Q1) and the 75th percentile (Q3) of the `value` column to determine the middle range of the data.

2. **Calculate IQR**:
$
   \text{IQR} = Q3 - Q1
$
   This represents the range between Q1 and Q3.

3. **Determine the boundaries for outliers**:
   - **Lower Bound**: $ Q1 - 1.5 \times \text{IQR} $
   - **Upper Bound**: $ Q3 + 1.5 \times \text{IQR} $
   - Values outside this range are considered outliers.

4. **Identify and count the outliers**: Find rows where the `value` column falls outside the bounds and count the number of outliers.

5. **Remove the outliers**: Create a new DataFrame `data_cleaned` after removing the rows containing outliers.


```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Calculate Q1 (25th percentile) and Q3 (75th percentile) for the 'Value' column
Q1 = data['Value'].quantile(0.25)
Q3 = data['Value'].quantile(0.75)

# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

# Calculate the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find rows that are outliers based on these bounds
outliers = (data['Value'] < lower_bound) | (data['Value'] > upper_bound)

# Count the number of outliers detected
print("Number of outliers detected in 'Value':", outliers.sum())

# Remove outliers from the DataFrame
data_cleaned = data[~outliers]

# Display information about the cleaned dataset
print("\nData cleaned. Shape of the new dataset:", data_cleaned.shape)
```
